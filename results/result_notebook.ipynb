```python
import json
with open("./binary_classifier_20_epochs_queer_results.json", "r") as f:
    results = json.load(f)
men_correct = []
men_incorrect = []
women_correct = []
women_incorrect = []
for gender in results.keys():
    for video in results[gender]:
        if video[0]["label"] == gender:
            if gender == "men":
                men_correct.append(video[0]["score"])

            else:
                women_correct.append(video[0]["score"])
        else:
            if gender == "men":
                men_incorrect.append(video[0]["score"])
            else:
                women_incorrect.append(video[0]["score"])

```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = {"men_correct": len(men_correct), "men_incorrect": len(men_incorrect), "women_correct": len(women_correct), "women_incorrect": len(women_incorrect)}
df = pd.DataFrame([df])
#define figure and axes
fig, ax = plt.subplots()

#hide the axes
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')

table = ax.table(cellText=df.values, colLabels=df.columns, loc='center')
#display table
fig.tight_layout()
plt.show()
```

```python
import numpy as np
df = {"men_correct": np.average(men_correct), "men_incorrect": np.average(men_incorrect), "women_correct": np.average(women_correct), "women_incorrect": np.average(women_incorrect)}
df = pd.DataFrame([df])
#define figure and axes
fig, ax = plt.subplots()

#hide the axes
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')

table = ax.table(cellText=df.values, colLabels=df.columns, loc='center')
#display table
fig.tight_layout()
plt.show()
```

```python
import json
with open("./tertiary_classifier_20_epochs_queer_results.json", "r") as f:
    results = json.load(f)
men_correct = []
men_incorrect = []
women_correct = []
women_incorrect = []

#for incorrect classification
men_women = []
women_men = []
men_other = []
women_other = []
for gender in results.keys():
    for video in results[gender]:
        if video[0]["label"] == gender:
            if gender == "men":
                men_correct.append(video[0]["score"])

            else:
                women_correct.append(video[0]["score"])
        else:
            if gender == "men":
                men_incorrect.append(video[0]["score"])
                if video[0]["label"] == "women":
                    men_women.append(video[0]["score"])
                elif video[0]["label"] == "other":
                    men_other.append(video[0]["score"])
            else:
                women_incorrect.append(video[0]["score"])
                if video[0]["label"] == "men":
                    women_men.append(video[0]["score"])
                elif video[0]["label"] == "other":
                    women_other.append(video[0]["score"])
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = {"men_correct": len(men_correct), "men_incorrect": len(men_incorrect), "women_correct": len(women_correct), "women_incorrect": len(women_incorrect)}
df = pd.DataFrame([df])
#define figure and axes
fig, ax = plt.subplots()

#hide the axes
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')

table = ax.table(cellText=df.values, colLabels=df.columns, loc='center')
#display table
fig.tight_layout()
plt.show()
```
